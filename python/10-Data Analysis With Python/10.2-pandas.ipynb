### Pandas-DataFrame And Series

Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning, it providers two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially hetrogeneous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
```

```python
## Series
# A Pandas Series is a one-dimensional array-like object that can hold any data type. It is similiar to column in a table.

data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n", series)
```

```python
# Create a Series from dictionary
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]
index = ['a','b','c']
pd.Series(data, index=index)
```

```python
# Data frame
#create a Dataframe from a dictionary of list

data= {
    'name':['Krish','John', 'Jack'],
    'Age':[25,30,45],
    'City':['Bangalore','New York', 'Florida']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
## Create a Data frame from a list of dictionaries

data=[
    {'Name': 'Krish', 'Age':32, 'city':'Bangalore'},
    {'Name': 'john', 'Age':34, 'city':'Bangalore'},
    {'Name': 'Bappy', 'Age':32, 'city':'Bangalore'},
    {'Name': 'Jack', 'Age':32, 'city':'Bangalore'},
]

df = pd.DataFrame(data)
print(df)
print(type(df))
```
