### Data Visualization With Matplotlib
Matplotlib is a powerful plotting library for Python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them.

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#create a line plot
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

# create a customized line plot
plt.plot(x,y, color='red', linestyle='--')
```

```python
plt.plot(x,y, color='red', linestyle='--', marker='o')
```

```python
plt.plot(x,y, color='red', linestyle='--', marker='o', linewidth=3, markersize=9)
```

```python
## Multiple Plots
## Sample data

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]
# y3=[10,8,6,4,2]

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(x,y1, color='green')
plt.title('plot 1')

plt.subplot(1,2,2)
plt.plot(y1,x, color='blue')
plt.title('plot 2')



# plt.plot(x,y, color='red',linestyle='dashed',linewidth=2,marker='o',markerfacecolor='blue',markersize=10)
```
